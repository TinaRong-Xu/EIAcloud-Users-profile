```python
class _trace:
    def __init__(self,read,ad,payment,chemical,xypt,siteinfocopy,download,WWWW,All):
        self.read=read
        self.ad=ad
        self.payment=payment
        self.chemical=chemical
        self.xypt=xypt
        self.siteinfocopy=siteinfocopy
        self.WWWW=WWWW
        self.download=download
        self.All=All
        
    def _getinfo(self):
        try:
        #WWWW=pd.DataFrame(columns=['手机号','日期','时间','设备号','地点：省','地点：市','事件类型','事件内容'])
            x=len(self.WWWW)
            for i in range(len(self.read)):
                #WWWW.append()
                row=[]
                row.append(self.read['username'].iloc[i])
                row.append(self.read['打开日期'].iloc[i])
                row.append(self.read['打开时间'].iloc[i])
                row.append(self.read['phoneId'].iloc[i])
                row.append(' ')
                row.append(' ')
                row.append(self.read['事件类型'].iloc[i])
                row.append(self.read['\ufeffnamecn_law'].iloc[i])
                WWWW.loc[x+i]=row
        except KeyError:
            pass        
            #print(x)
        self.WWWW=WWWW 
        
        try:
            x=len(self.WWWW)
            for i in range(len(self.ad)):
                #WWWW.append()
                row=[]
                row.append(self.ad['userId'].iloc[i])
                row.append(self.ad['打开日期'].iloc[i])
                row.append(self.ad['打开时间'].iloc[i])
                row.append(self.ad['androidId'].iloc[i])
                row.append(self.ad['province'].iloc[i])
                row.append(self.ad['city'].iloc[i])
                row.append(self.ad['事件类型'].iloc[i])
                row.append(self.ad['\ufeffAd'].iloc[i])
                WWWW.loc[x+i]=row
        except KeyError:
            pass        
            #print(x)
        self.WWWW=WWWW 
         
        try:
            x=len(self.WWWW)
            for i in range(len(self.payment)):
                #WWWW.append()
                row=[]
                row.append(self.payment['username'].iloc[i])
                row.append(self.payment['打开日期'].iloc[i])
                row.append(self.payment['打开时间'].iloc[i])
                row.append(' ')
                row.append(' ')
                row.append(' ')
                row.append(self.payment['事件类型'].iloc[i])
                if self.payment['pay_type'].iloc[i]==1:
                    self.payment['pay_type'].iloc[i]='微信'
                elif self.payment['pay_type'].iloc[i]==2:
                    self.payment['pay_type'].iloc[i]='支付宝'
                elif self.payment['pay_type'].iloc[i]==3:
                    self.payment['pay_type'].iloc[i]='IOS Applestore'
                elif self.payment['pay_type'].iloc[i]==4:
                    self.payment['pay_type'].iloc[i]='赠送'    
                y=str(self.payment['money'].iloc[i])+'元'+' '+str(self.payment['pay_type'].iloc[i])+'支付'+' '+str(self.payment['type'].iloc[i])
                row.append(y)
                WWWW.loc[x+i]=row
        except KeyError:
            pass        
            #print(x)
        self.WWWW=WWWW 

        
        try:    
            x=len(self.WWWW)
            for i in range(len(self.chemical)):
                #WWWW.append()
                row=[]
                row.append(self.chemical['username'].iloc[i])
                row.append(self.chemical['打开日期'].iloc[i])
                row.append(self.chemical['打开时间'].iloc[i])
                row.append(' ')
                row.append(' ')
                row.append(' ')
                row.append(self.chemical['事件类型'].iloc[i])
                row.append(self.chemical['law_name'].iloc[i])
                WWWW.loc[x+i]=row
        except KeyError:
            pass        
            #print(x)
        self.WWWW=WWWW 

        
        try:
            x=len(self.WWWW)
            for i in range(len(self.xypt)):
                #WWWW.append()
                row=[]
                row.append(self.xypt['username'].iloc[i])
                row.append(self.xypt['打开日期'].iloc[i])
                row.append(self.xypt['打开时间'].iloc[i])
                row.append(' ')
                row.append(' ')
                row.append(' ')
                row.append(self.xypt['事件类型'].iloc[i])
                row.append(self.xypt['law_name'].iloc[i])
                WWWW.loc[x+i]=row
        except KeyError:
            pass        
            #print(x)
        self.WWWW=WWWW 

       
        try:
            x=len(self.WWWW)
            for i in range(len(self.siteinfocopy)):
                #WWWW.append()
                row=[]
                row.append(self.siteinfocopy['telphone'].iloc[i])
                row.append(self.siteinfocopy['打开日期'].iloc[i])
                row.append(self.siteinfocopy['打开时间'].iloc[i])
                row.append(self.siteinfocopy['telephoneid'].iloc[i])
                row.append(self.siteinfocopy['\ufeff地区'].iloc[i])
                row.append(self.siteinfocopy['省'].iloc[i])
                row.append('登陆')
                row.append(' ')
                WWWW.loc[x+i]=row
        except KeyError:
            pass        
            #print(x)
        self.WWWW=WWWW 
    
        try:
            x=len(self.WWWW)
            for i in range(len(self.download)):
                #WWWW.append()
                row=[]
                row.append(self.download['username'].iloc[i])
                row.append(self.download['打开日期'].iloc[i])
                row.append(self.download['打开时间'].iloc[i])
                row.append(self.download['user_id'].iloc[i])
                row.append(' ')
                row.append(' ')
                row.append('下载文件')
                row.append(' ')
                WWWW.loc[x+i]=row
                #print('yes')
        except KeyError:
            pass        
            #print(x)
        self.WWWW=WWWW 
        return self.WWWW


        
    def _addinfo(self):
        #经常登陆时间段
        self.WWWW=self.WWWW.sort_values(by=['日期','时间']).reset_index(drop=True)
        #print(len(WWWW))
        self.WWWW.insert(3,'经常登陆时间段',0)
        for i in range(len(self.WWWW)):
            if str('07:00:00')<=str(self.WWWW['时间'].iloc[i])<=str('08:30:00'):
                                self.WWWW['经常登陆时间段'].iloc[i]='早通勤时间'
            elif str('08:30:00')<str(self.WWWW['时间'].iloc[i])<=str('11:30:00'):
                                self.WWWW['经常登陆时间段'].iloc[i]='上午工作时间'
            elif str('11:30:00')<str(self.WWWW['时间'].iloc[i])<=str('14:00:00'):
                                self.WWWW['经常登陆时间段'].iloc[i]='午休时间'
            elif str('14:00:00')<str(self.WWWW['时间'].iloc[i])<=str('18:00:00'):
                                self.WWWW['经常登陆时间段'].iloc[i]='下午工作时间'
            elif str('18:00:00')<str(self.WWWW['时间'].iloc[i])<=str('19:30:00'):
                                self.WWWW['经常登陆时间段'].iloc[i]='晚通勤时间'
            elif str('19:30:00')<str(self.WWWW['时间'].iloc[i])<str('24:00:00'):
                                self.WWWW['经常登陆时间段'].iloc[i]='夜晚'
            elif str('00:00:00')<=str(self.WWWW['时间'].iloc[i])<str('07:00:00'):
                                self.WWWW['经常登陆时间段'].iloc[i]='凌晨'  
        Time=self.WWWW.groupby('经常登陆时间段').count().reset_index()
        #print(self.WWWW)
        j=list(Time['手机号']).index(max(list(Time['手机号'])))
        All_time=Time.iloc[j]['经常登陆时间段']
        self.All_time=All_time
            
        #经常登陆地点
        Loc=self.WWWW.copy()
        Loc=Loc.groupby('地点：市').count().reset_index()
        j=list(Loc['手机号']).index(max(list(Loc['手机号'])))
        All_loc=Loc.iloc[j]['地点：市']
        #print(All_loc)
        Find=pd.DataFrame(columns=self.WWWW.columns)
        sum=0    
        for i in range(len(self.WWWW)):
            if self.WWWW['地点：市'].iloc[i]==All_loc:
                Find.loc[sum]=self.WWWW.iloc[i]
                sum+=1
        Find=Find.groupby('经常登陆时间段').count().reset_index()
        #print(Find)
        #print(Loc)
        #
        j=list(Find['手机号']).index(max(list(Find['手机号'])))
        #print(j)
        Tima_Loc=Time.iloc[j]['经常登陆时间段']
        #print(Tima_Loc)
        if All_loc==' ':
            All_loc='暂不可知'
        else:
            All_loc=All_loc
        if Tima_Loc=='上午工作时间' or Tima_Loc=='下午工作时间':
             All_loc=str('公司/项目所在地：')+All_loc
        elif Tima_Loc=='早通勤时间' or Tima_Loc=='晚通勤时间':
            All_loc=str('通勤途经地：')+All_loc
        elif Tima_Loc=='夜晚' or Tima_Loc=='凌晨':
             All_loc=str('家庭/临时住址所在地：')+All_loc
        elif Tima_Loc=='午休时间' :
             All_loc=str('午休所在地：')+All_loc

            
        self.All_time=All_time
        self.All_loc=All_loc
        
        
        #使用过的功能+经常使用的功能
        function=pd.DataFrame(self.WWWW.groupby('事件类型').count()).reset_index().sort_values(by='手机号',ascending=False)
        print(function)
        Func=list(function['事件类型'])
        if '登陆' in Func:
            Func.remove('登陆')
        #function_most=list(function['手机号']).index(max(list(function['手机号'])))
        Func_most=Func[0]
        self.All_Func=Func
        self.All_Func_most=Func_most
        print(Func)        
        
        #最后一次缴费时间及金额
        pay_time=self.payment.sort_values(by='打开日期',ascending=False)
        print(pay_time)
        try:
            Last_pay=str(pay_time.iloc[0]['打开日期'])+'|'+'支付'+str(pay_time.iloc[0]['money'])+'|'+str(pay_time.iloc[0]['type'])
        except IndexError:
            Last_pay='无缴费记录'
        self.All_Last_pay=Last_pay
            
            
        

        #曾用设备Iphone+安卓+小程序
        Dev_now=self.siteinfocopy.sort_values(by='打开日期')['telephoneid'].iloc[0]
        device=pd.DataFrame(self.siteinfocopy.groupby('telephoneid').count()).reset_index()
        sum_android=0
        sum_iphone=0
        sum_wechat=0
        #print(device)
        #print(device)
        for i in range(len(device)):
            try:
                #print(device['telephoneid'].iloc[i])
                c=len(device['telephoneid'].iloc[i])
                if c<32:
                    device['telphoneType'].iloc[i]='wechat'
                else:
                    if '-' in str(device['telephoneid'].iloc[i]):
                        device['telphoneType'].iloc[i]='ios'
                    else:
                        device['telphoneType'].iloc[i]='android'
            except IndexError:
                pass
        for i in range(len(device)): 
            try:     
                A=device.iloc[i]['telphoneType']
                if A=='android':
                    sum_android+=1
                elif A=='ios':
                    sum_iphone+=1
                elif A=='wechat':
                    sum_wechat+=1
            except IndexError:           
                pass
        #print(device)
        self.sum_android=sum_android
        self.sum_iphone=sum_iphone
        self.sum_wechat=sum_wechat
            
        self.WWWW.to_csv('test.csv',index=False)

        #最高日下载量
        Top=0
        try:
            top_download=pd.DataFrame(self.download.groupby('打开日期').count().reset_index()).sort_values(by='username',ascending=False)
            Top=top_download['username'].iloc[0]
        except IndexError:
            pass
        self.top=Top

             
        #行为积分
        times_day=pd.DataFrame(self.WWWW.groupby('日期').count()).sort_values(by='手机号',ascending=False)
        #print(times_day['手机号'])
        sumdate=0
        for i in range(len(times_day['手机号'])):
            y=int(times_day['手机号'].iloc[i])
            sumdate=sumdate+y
        Times_day=sumdate/len(times_day)
        print(Times_day)
        person=self.WWWW['手机号'].iloc[0]
        b=list(All['手机号']).index(person)
        x=int(len(self.WWWW)-1)
        #print(self.WWWW['日期'].iloc[x])
        #print(self.All['最高日下载量'].iloc['当前状态起始时间'])
        zhuangtai=self.All['当前状态起始时间'].apply(lambda x:x.split(' ')[0])
        for i in range(len(zhuangtai)):
            zhuangtai.iloc[i]=datetime.datetime.strptime(zhuangtai.iloc[i], '%d/%m/%Y').strftime('%Y-%m-%d')
        #print(zhuangtai[0])
        year1=self.WWWW['日期'].apply(lambda x:x.split('-')[0])
        month1=self.WWWW['日期'].apply(lambda x:x.split('-')[1])
        day1=self.WWWW['日期'].apply(lambda x:x.split('-')[2])
        date1=datetime.datetime(int(year1.iloc[x]),int(month1.iloc[x]),int(day1.iloc[x]))
        year2=zhuangtai.apply(lambda x:x.split('-')[0])
        month2=zhuangtai.apply(lambda x:x.split('-')[1])
        day2=zhuangtai.apply(lambda x:x.split('-')[2])    
        date2=datetime.datetime(int(year2.iloc[b]),int(month2.iloc[b]),int(day2.iloc[b]))
        #print(date1)
        #print(date2)
        date=date1-date2
        date=int(date.days)
        score=len(self.WWWW)*len(Func)*Times_day/date
        score="%.2f" % score
        self.score=score
        return self.All_time,self.All_loc,self.All_Func,self.All_Func_most,self.All_Last_pay,self.sum_android,self.sum_iphone,self.sum_wechat,self.top,self.score


    def _write(self):
        self._getinfo()
        self._addinfo()
        person=self.WWWW['手机号'].iloc[0]
        print(person)
        b=list(All['手机号']).index(person)
        self.All['经常登陆时间段'].iloc[b]=self.All_time
        self.All['常用地址'].iloc[b]=self.All_loc
        self.All['使用过的功能'].iloc[b]=self.All_Func
        self.All['经常使用的功能'].iloc[b]=self.All_Func_most
        self.All['最后一次缴费时间及金额'].iloc[b]=self.All_Last_pay
        self.All['曾用设备：安卓'].iloc[b]=self.sum_android
        self.All['曾用设备：Iphone'].iloc[b]=self.sum_iphone
        self.All['曾用设备：小程序'].iloc[b]=self.sum_wechat
        self.All['最高日下载量'].iloc[b]=self.top
        self.All['使用水平积分'].iloc[b]=self.score
        self.All.to_csv('/Users/rong/Desktop/尚云/客户分析/表2/30人当前状态.csv',encoding='utf_8_sig',index=False)
        return self.All.iloc[b]
```
